# original spectra-corn dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from process import msc, snv, detrend, derivative, savitzky_golay, airPLS
plt.rcParams.update({'font.size': 16})

data_path = 'D:/A/CSU/corn_xlsl/m5_corn.xlsx'
df = pd.read_excel(data_path)
spectra = df.iloc[:, :700].values 

wavelengths = np.arange(1100, 1100 + 2 * 700, 2)
plt.figure(figsize=(10, 5))
for i in range(spectra.shape[0]):  
    plt.plot(wavelengths, spectra[i], color='blue', alpha=0.3, label='orgin' if i == 0 else "")
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance (AU)')
plt.title('Original Spectra')
plt.legend()
plt.tight_layout()
plt.show()
```
